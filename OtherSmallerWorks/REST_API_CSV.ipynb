Here is a code to show my learning of REST API Integration. I'll fetch the latest COVID-19 data from a reliable API. I'm using the COVID-19 Data Repository by the Center for Systems Science and Engineering (CSSE) at Johns Hopkins University.

```python
# In this project I'm using the request.get() method to send a GET request to the URL of the COVID-19 data.

import requests
import pandas as pd

url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'

def fetch_latest_covid_data(url):
    
    
    response = requests.get(url)
    
    # If the response status code is 200 request is successful
    if response.status_code == 200:
        return response.content
    else:
        print(f"Error: Unable to fetch data. Status code: {response.status_code}")
        return None

covid_data = fetch_latest_covid_data(url)

# Saving the data to a file
if covid_data:
    with open('latest_covid19_data.csv', 'wb') as file:
        file.write(covid_data)
        print("Data saved successfully.")
```
